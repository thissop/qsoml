```python
import pandas as pd 
from astropy.io import fits 
import matplotlib.pyplot as plt 
import smplotlib 
import numpy as np
import os 

data_path = '/Users/tkiker/Documents/GitHub/qsoml/data/small-batch/spec-3587-55182-0346.fits'

with fits.open(data_path) as hdul:
    data = hdul[1].data
    wave = 10**np.array(data['loglam'])
    ivar = np.array(data['ivar'])
    model_flux = np.array(data['model'])
    flux = np.array(data['flux'])

    fix, axs = plt.subplots(2, 1, figsize=(6, 6))

    axs[0].plot(wave, flux, label='Raw')
    axs[0].plot(wave, model_flux, color='red', label='Model')
    axs[0].legend(fontsize='small')

    axs[1].plot(wave, 1/ivar, label='Raw (ivar)')
    axs[1].plot(wave, model_flux, color='red', label='Model')
    axs[1].legend(fontsize='small')

    plt.show()
```
